```python
new_dataset = pd.read_csv('data.csv')
```

```python
new_dataset.info()
```

```python
new_dataset.drop(columns=['property_id', 'location_id', 'page_url', 'location',
                 'city', 'province_name', 'purpose', 'date_added',
                 'agency', 'agent'], inplace=True)
```

```python
new_dataset.info()
```

```python
new_dataset.head()
```

```python
new_dataset['Area Category'].unique()
```

```python
# Define the "maximum allowed" categories
allowed_categories = [
    '0-5 Marla',
    '5-10 Marla',
    '1-5 Kanal',
    '10-15 Marla',
    '5-10 Kanal'
]

# Keep only allowed categories and drop the rest
new_dataset = new_dataset[new_dataset['Area Category'].isin(allowed_categories)].copy()

# Verify
print(new_dataset['Area Category'].unique())
print(new_dataset.shape)
```

```python
new_dataset.info()
```

```python
new_dataset.head()
```

```python
corrs = new_dataset.corr(numeric_only=True)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(
    corrs,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5
)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
def area_to_numeric(area_str):
    try:
        # Remove commas and strip spaces
        value, unit = area_str.replace(',', '').split()
        value = float(value)
        if unit.lower() == 'marla':
            return value
        elif unit.lower() == 'kanal':
            return value * 20  # 1 Kanal = 20 Marla
        else:
            return None
    except:
        return None  # in case of missing or malformed values
```

```python
# Apply to dataset
new_dataset['Area Size (Marla)'] = new_dataset['area'].apply(area_to_numeric)
```

```python
new_dataset['area'].unique()
```

```python
# List of huge values to drop
new_dataset = new_dataset[~new_dataset['area'].isin(['4,450 Kanal', '4,500 Kanal', '6,222.2 Kanal', '1,000 Kanal', '2,250 Kanal'])].copy()
```

```python
new_dataset = new_dataset[~new_dataset['area'].isin(['1,722.2 Kanal'])].copy()
```
